# Ilustrasi Membuat Harmonograph Dengan Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter yang digunakan
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)

# Silakan merubah parameter ini
A = [  10, 15, 8, 9 ] # Amplitudo
d = [ .004, .001, .002, .0015 ] # Dumping
f = [   3, 1, 4, 2.5 ] # Frekuensi

# Membuat pasangan x dan y (Formula membuat homograph)
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# Menampilkan plot nya
plt.plot(x,y,'b',linewidth=.1)
plt.axis('off')
plt.show()
```

```python
# Parameter yang digunakan
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)

# Silakan merubah parameter ini
A = [  2,2,5,5,]
d = [ .0008, .0009, .0001, .0002] # Berkaitan dengan Volume
f = [  2,4, 8, 10] 

# Membuat pasangan x dan y
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# Menampilkan plot nya
plt.plot(x,y,'b',linewidth=.1)
plt.axis('off')
plt.show()
```

```python
# Parameter yang digunakan
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)

# Silakan merubah parameter ini
A = [  1, 1, 1.5, 1.5 ]
d = [ .004, .001, .002, .0015 ]
f = [   3, 3, 4, 4 ]

# Membuat pasangan x dan y
x = A[0]*np.sin(t*f[0]+90/2)*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1]+90/2)*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# Menampilkan plot nya
plt.plot(x,y,'r',linewidth=.1)
plt.axis('off')
plt.show()
```
